This notebook demonstrates how to visualize distributions using Seaborn and Matplotlib. We'll focus on creating intuitive and beginner-friendly plots that help understand the data better.

### Importing Libraries

We begin by importing the necessary Python libraries for data visualization.

### Loading the Dataset

Next, we load a built-in dataset using Seaborn. The `tips` dataset contains information about restaurant bills and tips received, which is great for visual exploration.

### Visualizing Distribution with Histograms and KDE

We'll start by plotting the distribution of numerical data. Histograms show the frequency of values, and Kernel Density Estimation (KDE) provides a smoothed curve to understand the shape of the distribution.

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

### Customizing the Plot Appearance

It's important to customize plots for clarity. We'll add titles, labels, and improve aesthetics to make the visualizations more readable and engaging.

We use distribution plots to visualize the distribution of quantitative data.


Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

* distplot/histplot
* jointplot
* pairplot
* rugplot

### Plotting Subsets of Data

To gain more insight, we can break down the data by categoriesâ€”like genderâ€”and plot distributions separately.

```python
df = sns.load_dataset('tips')
df
```

### Conclusion

Visualizing data helps uncover patterns and trends that are not obvious from raw numbers. By combining histograms and KDE plots, we can better understand data distributions.

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)

```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'reg')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
sns.rugplot(df['tip'])
```
